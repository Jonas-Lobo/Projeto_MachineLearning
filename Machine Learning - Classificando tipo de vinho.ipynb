```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
from sklearn.ensemble import ExtraTreesClassifier
```

```python
df = pd.read_csv('wine_dataset.csv')
```

```python
df.head()
```

```python
#Transformando as variaveis de texto para numericos
df['style'] = df['style'].replace('red', 0)
df['style'] = df['style'].replace('white', 1)
```

```python
df.head()
```

```python
#Separando as variaveis que serão usadas para prever e outras para alvo
y = df['style']
x = df.drop('style', axis = 1)
```

```python
#Pegando uma parte do dataset para treino e teste
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.3)
```

```python
#Criando modelo
modelo = ExtraTreesClassifier()
modelo.fit(x_treino, y_treino)

#Imrimindo resultado
resultado = modelo.score(x_teste, y_teste)
print('Acurácia:', resultado)
```
